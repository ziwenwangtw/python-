sin wave

```python
import matplotlib.pyplot as pt
import numpy as np
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
z=np.cos(x*np.pi/180.0)
pt.xlim(0,360)
pt.ylim(-1.2,1.2)
pt.plot(x,y,label="sin")
pt.plot(x,z,label="cos")
pt.title("Sin & Cos wave")
pt.xlabel("Degree")
pt.ylabel("Value")
pt.legend()
pt.show()
```

3D圖

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
from matplotlib import cm

def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

xgrid = np.linspace(-3, 3, 50)
ygrid = xgrid 
x, y = np.meshgrid(xgrid, ygrid)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, f(x, y), rstride=2, cstride=2, cmap=cm.jet, alpha=0.7, linewidth=0.25)
ax.set_zlim(-0.5, 1.0)
plt.show()
```

k-means群聚clustering演算法

```python
from sklearn import datasets, cluster, metrics
import matplotlib.pyplot as pt
iris=datasets.load_iris()
iris_data=iris.data
#執行k-means群聚演算法
ks=range(2,11)
silhouette_avgs=[]
for n in ks:
    result=cluster.KMeans(n_clusters=n).fit(iris_data)
    #resutl.labels_ 預測結的分群結果
    #print(result.labels_)
    #用silhouette score評估分群果好壞，越接近1越好
    silhouette_avg=metrics.silhouette_score(iris_data, result.labels_)
    silhouette_avgs.append(silhouette_avg)
#print(silhouette_avg)
pt.bar(ks, silhouette_avgs)
pt.show()
```

```python
from sklearn import datasets, svm
import matplotlib.pyplot as plt
digits=datasets.load_digits()
from sklearn.cross_validation import train_test_split 
X_train, X_test, y_train, y_test, images_train, images_test = train_test_split(digits.data, digits.target, digits.images, test_size=0.25, random_state=42)

svc_model = svm.SVC(gamma=0.001, C=100., kernel='linear')
svc_model.fit(X_train, y_train)
predicted = svc_model.predict(X_test)
images_and_predictions = list(zip(images_test, predicted))

for index, (image, prediction) in enumerate(images_and_predictions[:5]):
    plt.subplot(1, 5, index + 1)
    plt.axis('off')
    plt.imshow(image, cmap=plt.cm.binary)
    plt.title('Predicted: ' + str(prediction))
plt.show()
```
